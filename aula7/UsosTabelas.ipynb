## Exemplo 2

Considere a tabela  com as visualizações de um usuário de Netflix. Qual a série mais vista por esse usuário?

```python
netflix = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/netflix.csv')
netflix
```

É possível notar olhando a coluna Title da tabela, que no caso das séries, é especificado a Temporada e o nome do episódio. Se fossemos achar a moda (valor mais frequente) da coluna Title, ele não daria como resultado a série mais vista. Para isso, precisamos criar uma nova coluna com o nome da série (ou filme no caso de filme). Para isso, usa-se a função str.split para separar o conteúdo da coluna. É possível observar que o caracter : é usado para separar o nome da série do nome da coluna. O resultado da separação será gravado na coluna Programa.

```python
netflix['Programa'] = netflix['Title'].str.split(':', n=1, expand=True)[0]
netflix
```

Agora é possível usar a funçáo mode para calcular a moda da coluna Programa.

```python
netflix['Programa'].mode()
```

## Exercício 3
Considere a tabela  contendo os dados do Campeonato Brasileiro de 2018, que está em https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/tabelaBrasileirao2018.csv.


Escreva um código que :

1) Leia a tabela e receba o nome de um time que participou do campeonato.

2) Receba uma opção: saldo, soma, tomados

3) Se no passo 2) foi digitado digitou:

- saldo: calcular e imprimir o valor do saldo de gols do time no campeonato.

- soma: calcular e imprimir o valor total de gols feitos pelo time no campeonato.

- tomados: calcular e imprimir o valor total de gols tomados pelo time no campeonato.

- Se o time não participou do campeonato ou não foi digitado nenhuma das opções do item 2: imprimir mensagem "Erro na entrada"

```python
tabela = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/tabelaBrasileirao2018.csv')
tabela
```

Uma possível solução está abaixo. Para verificar se o time está é um time foi usado o comando `in`. Ele retorna `True` se estiver na lista e `False` se não estiver.

```python
time = input('Digite um time:')


if time in tabela['Mandante'].unique():
    opcao = input('Digite uma opção [saldo, soma, tomados]:')
    if opcao == 'saldo':
        saldo_mandante = tabela.query('Mandante == "'+time+'"')['Placar do Mandante'] - tabela.query('Mandante == "'+time+'"')['Placar do Visitante']
        saldo_visitante = tabela.query('Visitante == "'+time+'"')['Placar do Visitante'] - tabela.query('Visitante == "'+time+'"')['Placar do Mandante']
        saldo = saldo_mandante.sum() + saldo_visitante.sum()
        print('O time', time, 'teve um saldo de', saldo, 'gols')
    elif opcao == 'soma':
        soma_mandante = tabela.query('Mandante == "'+time+'"')['Placar do Mandante']
        soma_visitante = tabela.query('Visitante == "'+time+'"')['Placar do Visitante']
        soma = soma_mandante.sum() + soma_visitante.sum()
        print('O time', time, 'teve um total de gols igual a', soma)
    elif opcao == 'tomados':
        tomado_mandante = tabela.query('Mandante == "'+time+'"')['Placar do Visitante']
        tomado_visitante = tabela.query('Visitante == "'+time+'"')['Placar do Mandante']
        tomado = tomado_mandante.sum() + tomado_visitante.sum()
        print('O time', time, 'levou um total de', tomado, 'gols')
    else:
        print('Erro de entrada')
else:
    print('Erro de entrada')
    
    
```

## Exercício 4

Considere a tabela com diversas informações de sujeitos que participaram de um experimento de controle da postura, encontrada em https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/BDSinfo.csv.

O seu código deve:

1) Ler o link da planilha e receber um número do sujeito (coluna Subject) a ser considerado.

2) Mostrar se a altura desse sujeito é maior, menor ou igual à média de altura dos sujeitos do mesmo Gênero.

```python
data = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/BDSinfo.csv', sep='\t')
data
```

Uma solução está mostrada abaixo. Para pegar os dados de uma linha de um DataFrame do Pandas, uas-se o comando `iloc[n]`, em que `n` é a linha desejada.

```python
sujeito = input('Digite o número de um sujeito:')
genero = data.query('Subject == ' + sujeito)['Gender'].iloc[0]
altura = data.query('Subject == ' + sujeito)['Height'].iloc[0]

media_altura_genero =  data.query('Gender == "' + genero + '"')['Height'].mean()

if altura > media_altura_genero:
    print('O sujeito', sujeito, 'tem a altura igual a', altura, 'cm, maior que a média de altura do gênero', genero)
elif altura < media_altura_genero:
    print('O sujeito', sujeito, 'tem a altura igual a', altura, 'cm, menor que a média de altura do gênero', genero)
else:
    print('O sujeito', sujeito, 'tem a altura igual a', altura, 'cm, igual que a média de altura do gênero', genero)
```

## Exemplo 5

Considere a planilha que contém a porcetagem de analfabetismo nas Unidades da Federação do Brasil de 1981 a 2014, encontrada no link https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/analfabetismo.csv .

Faça um programa contendo instruções para ler a planilha e 

1) receber do teclado uma Unidade da Federação.

2) calcular a regressão linear entre os anos e a taxa de analfabetismo da Unidade da Federação escolhida.

3) fazer o gráfico de dispersão da taxa de analfabetismo da Unidade da Federação. No mesmo gráfico, fazer a reta da regressão linear encontrada.

4) receber do teclado um ano maior do que 2014 e mostrar a previsão da taxa de analfabetismo para a Unidade da Federação escolhida.

5) Caso não seja digitado nenhuma Unidade da Federação ou um ano menor ou igual a 2014, mostrar a mensagem `Erro de entrada`.






```python
analfabetismo = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/analfabetismo.csv')
analfabetismo
```

Uma solução está abaixo. Para listar o nome das unidades da federação, foi utilizada a função `.columns()`, que lista o nome de todas as colunas de uma tabela. 

```python
uf = input('Digite uma unidade da Federação:')
if uf in analfabetismo.columns:
    m, b = np.polyfit(analfabetismo['Ano'], analfabetismo[uf], deg=1)

    plt.figure()
    plt.plot(analfabetismo['Ano'], analfabetismo[uf], marker='o', linestyle='')
    plt.plot(analfabetismo['Ano'], analfabetismo['Ano']*m+b)
    plt.show()

    ano = int(input('Digitar um ano maior do que 2014:'))

    prev = ano*m + b
    if ano > 2014:
        print('No ano de', ano ,' a previsão é que a taxa de analfabetismo em', uf, 'seja %.2f %%.' %prev)
    else:
        print('Erro de entrada')
else:
    print('Erro de entrada')
```

## Exemplo 6

Considere a tabela com diversas informações de sujeitos que participaram de um experimento de controle da postura, encontrada em https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/BDSinfo.csv.

Pede-se:

1) Ler a tabela e pedir para ser digitado duas colunas da tabela.
2) Se as duas colunas forem numéricas (tipo float64 ou int64), calcular o índice de correlação de cada uma das colunas com a coluna `Falls12m`. Essa coluna contém o número de quedas nos 12 meses que antecederam o experimento.
3) Mostrar na tela qual das colunas tem um módulo da correlação com  a coluna `Falls12m` maior e as duas correlações.

```python
data = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais2021/main/dados/BDSinfo.csv', sep='\t')
data
```

Uma possível solução é mostrada abaixo:

```python
col1 = input('Digite o nome de uma coluna:')
col2 = input('Digite o nome de uma coluna (diferente da primeira coluna):')

col1_numerica = data[col1].dtype == 'float64' or data[col1].dtype == 'int64'
col2_numerica = data[col2].dtype == 'float64' or data[col2].dtype == 'int64'


if col1_numerica and col2_numerica and col1 != col2:
    corr1 = data[col1].corr(data['Falls12m'])
    corr2 = data[col2].corr(data['Falls12m'])
    if np.abs(corr1) > np.abs(corr2):
        print(col1, 'tem o módulo da correlação (%.3f' %corr1,') com Falls12m maior do que', col2, '(%.3
              f' %corr2 ,')')
    elif np.abs(corr2) > np.abs(corr1):
        print(col2, 'tem o módulo da correlação (%.3f' %corr2,') com Falls12m maior do que', col1, '(%.3f' %corr1 ,')')
    else:
        print(col1, 'tem o módulo da correlação (%.3f' %corr1,') com Falls12m igual a', col2, '(%.3f' %corr2 ,')')
else:
    print('Erro de entrada')
```
