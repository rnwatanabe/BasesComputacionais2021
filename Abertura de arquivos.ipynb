# Dicas de como abrir alguns tipos de arquivos no Pandas

```python
import pandas as pd
```

## Arquivos Excel (extensão .xls, .xlsx, .ods)

Para abrir arquivos Excel, basta usar a função do Pandas pd.read_excel().  Por exemplo, no arquivo abaixo do IBGE:

```python
dado = pd.read_excel('https://ftp.ibge.gov.br/Retroprojecao_da_populacao/2010_2000/retroprojecao_2018_indicadores_20180726.xls', skiprows=3)
dado
```

Às vezes, com arquivos desse tipo aparece o erro 403, como nessa situação com um arquivo do Ministério da Justiça:

```python
dado = pd.read_excel('http://dados.mj.gov.br/dataset/a4b1cf24-3ba7-4cf9-8fde-d2362af4aacc/resource/260622fe-4c07-4a91-a7c8-1efbadf767ce/download/extracaotabnatativasinativas20190917.xlsx')
dado
```

Para contornar a situação, pode-se usar a seguinte estrutura de código:

```python
from urllib.request import Request, urlopen 
url = 'http://dados.mj.gov.br/dataset/a4b1cf24-3ba7-4cf9-8fde-d2362af4aacc/resource/260622fe-4c07-4a91-a7c8-1efbadf767ce/download/extracaotabnatativasinativas20190917.xlsx' #colocar aqui o endereço do arquivo
req = Request(url)
req.add_header('User-Agent', 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0')
content = urlopen(req)

dado = pd.read_excel(content.read(), sheet_name=None, dtype='object')
```

Essa planilha Excel tem várias abas. Para ver quais abas existem, faça:

```python
abas = dado.keys()
print(abas)
```

Para mostrar a primeira aba:

```python
dado['listaNatureza']
```

## Arquivos armazenados no Google Drive

Para usar no Pandas um arquivo armazenado no Google Drive, você deve obter o link para download do arquivo. Por exemplo, o arquivo em [https://drive.google.com/file/d/1c32oVSlbZkPhT1WdT-ECHDYzvK1iuAaG/view?usp=sharing](https://drive.google.com/file/d/1c32oVSlbZkPhT1WdT-ECHDYzvK1iuAaG/view?usp=sharing). Na página que abre, no canto direito superior tem 3 pontos na vertical. Apertando esse botão, aparece a opção "Fazer Download". Isso faz abrir uma caixa de diálogo para salvar o arquivo localmente. Antes dessa caixa de diálogo abrir, aperte a tecla ESC do teclado. Isso faz com que a caixa de diálogo não abra. O endereço que aparece na barra de endereço da barra de endereços do navegador Web é o link para usar no Python. O link de exemplo dá o link [https://drive.google.com/u/1/uc?id=1c32oVSlbZkPhT1WdT-ECHDYzvK1iuAaG&export=download](https://drive.google.com/u/1/uc?id=1c32oVSlbZkPhT1WdT-ECHDYzvK1iuAaG&export=download). Como esse arquivo é csv (com delimitador uma tabulação '\t'), para abri-lo, basta:



```python
dado = pd.read_csv('https://drive.google.com/u/1/uc?id=1c32oVSlbZkPhT1WdT-ECHDYzvK1iuAaG&export=download', sep='\t')
dado
```
